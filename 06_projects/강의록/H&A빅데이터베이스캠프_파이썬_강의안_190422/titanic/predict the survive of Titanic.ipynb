#타이타닉 생존자 예측 

```python
import pandas as pd
```

```python
train = pd.read_csv("train.csv", index_col = ["PassengerId"])
train.head()
train.shape
```

```python
test = pd.read_csv("test.csv", index_col = ["PassengerId"])
test.head()
test.shape
```

## 데이터 전처리

- 성별 인코딩 

```python
train.loc[train["Sex"]=="male", "Sex_encode"] = 0
train.loc[train["Sex"]=="female", "Sex_encode"] = 1
train.head()
```

```python
test.loc[test["Sex"]=="male", "Sex_encode"] = 0
test.loc[test["Sex"]=="female", "Sex_encode"] = 1
test.head()
```

- 결측치 채우기

```python
test[pd.isnull(test['Fare'])]
```

```python
mean_fare = train['Fare'].mean()
mean_fare
```

```python
test.loc[pd.isnull(test['Fare']), 'Fare'] = mean_fare
```

- 호칭 컬럼 만들기

```python
def find_title(name):
    return name.split(', ')[1].split('.')[0]
```

```python
train["Title"] = train["Name"].apply(find_title)
train.head(10)
```

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.countplot(data=train, x='Title', hue = 'Survived')
```

- Embarked One-Hot Encoding

```python
train_embarked = pd.get_dummies(train["Embarked"])
train_embarked.head()
```

```python
train = pd.concat([train, train_embarked], axis=1)
train.head()
#train
```

```python
test_embarked = pd.get_dummies(test["Embarked"])
test_embarked.head()

test = pd.concat([test, test_embarked], axis=1)
test.head()
```

```python
sns.countplot(data = train, x = 'SibSp', hue = 'Survived')
```

```python
sns.distplot(train['Fare'])
```

```python
ignore_high_fare = train[train['Fare'] < 100]

perish = ignore_high_fare[ignore_high_fare['Survived'] == 0]
survived = ignore_high_fare[ignore_high_fare['Survived'] == 1]

sns.distplot(perish['Fare'], label = 'Perish')
sns.distplot(survived['Fare'], label = 'Survived')
```

# 학습시키기

```python
feature_names = ['Pclass', 'Sex_encode', 'Fare', 'C', 'Q', 'S']
X_train = train[feature_names]
Y_train = train['Survived']

X_test = test[feature_names]
```

## 모델선택

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 3000, max_depth = 5)
```

## 학습

```python
model.fit(X_train, Y_train)
```

```python
prediction = model.predict(X_test)
prediction
```

# Submission

```python
submission = pd.read_csv("gender_Submission.csv", index_col = ['PassengerId'])

submission['Survived'] = prediction
submission.head()

submission.to_csv('my_first_ML.csv')
```
