```python
################################################### 여러개의 CSV 파일 가져오기

import glob
import os
import sys

input_path ="" # 폴더위치 지정
output_file = "Combined Sales.csv" #합친파일이름

all_files = glob.glob(os.path.join('sales_*'))

#print(all_files)
#print(all_files[0])

all_data_frames = []
for i in all_files:
    data = pd.read_csv(i, index_col = None)
    all_data_frames.append(data)

#print(all_data_frames)
    
data_concat = pd.concat(all_data_frames, axis = 0, ignore_index = True)
print(data_concat)
data_concat.to_csv(output_file, index = False)
```

```python
### 각각의 파일에서 데이터 값의 합계 및 평균 구하기

all_files = glob.glob(os.path.join('sales_*'))

all_data_frames = []
for i in all_files:
    data = pd.read_csv(i, index_col = None)
    
    total_sales = pd.DataFrame([float(str(value).strip('$').replace(',',''))
                               for value in data.loc[:,'Sale Amount']]).sum()
    average_sales = pd.DataFrame([float(str(value).strip('$').replace(',',''))
                               for value in data.loc[:,'Sale Amount']]).mean()
    
    dic = {'file_name' : os.path.basename(i), 'total_sales' : total_sales, 'average_sales' : average_sales}
    all_data_frames.append(pd.DataFrame(dic, columns=['file_name', 'total_sales', 'average_sales']))
    
data_concat = pd.concat(all_data_frames, axis = 0, ignore_index = True)

data_concat
```
