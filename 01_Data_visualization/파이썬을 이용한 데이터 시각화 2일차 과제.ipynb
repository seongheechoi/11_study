## 다음과 같은 데이터를 사용하여 답하시오.
데이터1과 데이터2를 아래와 같이 생성한다.

```python
data1 = pd.DataFrame({'이름' : ['유관순', '유관순', '황진이', '황진이', '유관순'],
                     '중간' : [10, 20, 30, 40, 50]})
data2 = pd.DataFrame({'이름' : ['황진이', '유관순', '신사임당'],
                     '기말' : [30, 40, 50]})
data2
```

데이터1과 데이터2를 합쳐서 결과1과 같이 생성

```python
result1 = pd.merge(data1, data2, on='이름', how='inner')
result1
```

데이터1과 데이터2를 합쳐서 결과2과 같이 생성

```python
result2 = pd.merge(data1, data2, on='이름', how='outer')
result2
```
