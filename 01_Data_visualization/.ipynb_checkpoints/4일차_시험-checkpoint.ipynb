### 1. unemployment_rate.csv 파일을 이용하여 아래 그림과 같이 선 그래프를 그리시오

```python
df = pd.read_csv('unemployment_rate.csv', encoding='cp949', index_col='연도')
df = df.loc[:'2005년', :].T
print(df)
print(df.index)
plt.figure()
plt.plot(df.index, df['2001년'], marker='o', ms=4, label='2001년')
plt.plot(df.index, df['2002년'], marker='o', ms=4, label='2002년')
plt.plot(df.index, df['2003년'], marker='o', ms=4, label='2003년')
plt.plot(df.index, df['2004년'], marker='o', ms=4, label='2004년')
plt.plot(df.index, df['2005년'], marker='o', ms=4, label='2005년')
plt.legend(title='연도')
plt.xlabel('년령대')
plt.ylabel('연도명')
plt.title('년령대별 연도명 꺽은 선')

```

### 2. human_info.csv 파일을 이용하여 아래의 조건에 맞게 구상후 아래에 나열된 그래프를 그리시오

```python
df2 = pd.read_csv('human_info.csv', encoding='cp949')
df2
```

연령대 데이터 추가

```python
def newAge(x):
    if x < 30:
        return '청년'
    elif x >=30 and x < 60:
        return '중년'
    else:
        return '노년'
    
df2['연령대'] = df2['age'].apply(newAge)
df2
```

```python
import seaborn as sns
```

```python
plt.figure()
sns.countplot(x='marriage', data=df2, order=['결혼', '이혼', '미혼'])
plt.title('결혼 유무에 따른 빈도')
```

```python
plt.figure()
sns.countplot(x='marriage', data=df2, hue='religion', order=['결혼', '이혼', '미혼'])
plt.title('결혼 유무에 따른 빈도')
```

```python
x= df2['age']
plt.figure()
sns.distplot(x, hist=True, kde=True)
plt.title('kde와 histogram')
```

```python
pivot = df2.pivot_table(index='gender', columns='marriage', values='age')
plt.figure()
sns.heatmap(data=pivot, annot=True)
plt.title('결혼 유무와 성별에 대한 히트맵')
```

```python
cor = df2.corr()
plt.figure()
sns.heatmap(data=cor, annot=True, cmap='YlGnBu')
```
